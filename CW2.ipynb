# Data Generation

```python
import numpy as np
import matplotlib.pyplot as plt
import math
import tqdm

N = 25
X = np.reshape(np.linspace(0, 0.9, N), (N, 1))
Y = np.cos(10*X**2) + 0.1 * np.sin(100*X)
plt.scatter(X,Y,color='darkred')

y = np.matrix(Y)

x_pred = np.linspace(-0.3, 1.3, 200)
```

# Linear Regression - Polynomial Functions

```python
def phi(x,order):
    return(x**order)
```

Order 0 corresponds to M = 1.

```python
M = 1
phi0 = []
for i in range(len(X)):
    line = []
    for m in range(M):
        line += [phi(X[i][0],m)]
    phi0 += [line]
        
phi0 = np.matrix(phi0)

phi0_pred = []
for i in range(len(x_pred)):
    line = []
    for m in range(M):
        line += [phi(x_pred[i],m)]
    phi0_pred += [line]
        
phi0_pred = np.matrix(phi0_pred)
```

```python
inv0 = np.linalg.inv(np.dot(phi0.T,phi0))
weight0 = np.dot(np.dot(inv0,phi0.T),y)
weight0
```

```python
mean0 = np.dot(weight0.T,phi0_pred.T)
```

```python
plt.scatter(X,Y,color='darkred',label = 'data')
plt.plot(x_pred,mean0.T,label = 'Order 0')
plt.title('Regression with Polynomial basis functions')
plt.legend()
```

# Linear Regression - Trigonometric Functions

```python
def trigo(x,order):
    if order == 0:
        return(1.0)
    elif order%2 == 0 :
        return(np.cos(math.pi*order*x))
    elif order%2 == 1 :
        return(np.sin(math.pi*(order+1)*x))
```

Order 1 corresponds to M = 3.

```python
M = 3
trigo1 = []
for i in range(len(X)):
    line = []
    for m in range(M):
        line += [trigo(X[i][0],m)]
    trigo1 += [line]
        
trigo1 = np.matrix(trigo1)

trigo1_pred = []
for i in range(len(x_pred)):
    line = []
    for m in range(M):
        line += [trigo(x_pred[i],m)]
    trigo1_pred += [line]
        
trigo1_pred = np.matrix(trigo1_pred)
```

```python
mean1 = np.dot(weight1.T,trigo1_pred.T)

```

# Over-Fitting - Leave One Out  Cross Validation

```python
training_error_array = []
testing_error_array = []
orders = []
for order in range(11):
    orders+=[order]
    M = 2*order + 1
    training_error_order = 0
    testing_error_order = 0
    for one_out in range(25):
        X_train = np.reshape(list(np.linspace(0, 0.9, N)[0:one_out])+list(np.linspace(0, 0.9, N)[one_out+1:N]), (N-1, 1))
        X_test = np.reshape(X[one_out], 1)
        y_train = np.cos(10*X_train**2) + 0.1 * np.sin(100*X_train)
        y_test = np.cos(10*X_test**2) + 0.1 * np.sin(100*X_test)


        trigo1 = []
        for i in range(len(X_train)):
            line = []
            for m in range(M):
                line += [trigo(X_train[i][0],m)]
            trigo1 += [line]
        
        trigo1 = np.matrix(trigo1)

        trigo1_pred = []
        for i in range(len(X)):
            line = []
            for m in range(M):
                line += [trigo(X[i][0],m)]
            trigo1_pred += [line]
        
        trigo1_pred = np.matrix(trigo1_pred)

        inv1 = np.linalg.inv(np.dot(trigo1.T,trigo1))
        weight1 = np.dot(np.dot(inv1,trigo1.T),y_train)
        mean1 = np.array(np.dot(weight1.T,trigo1_pred.T).T)
        training_error = 0
        testing_error = 0
        for i in range(len(mean1)):
            if i == one_out:
                testing_error += (mean1[i][0]-np.array(y)[i][0])**2
            else:
                training_error += (mean1[i][0]-np.array(y)[i][0])**2
        training_error = training_error/24.0
        training_error_order += training_error/25.0
        testing_error_order += testing_error/25.0
    
    training_error_array += [training_error_order]
    testing_error_array += [testing_error_order]
```

```python
plt.figure(3, figsize=(16, 10))
plt.title('Influence of Order')
plt.plot(orders,training_error_array,label = 'Training Error')
plt.plot(orders,testing_error_array,label = 'Testing Error')
plt.legend()
```

# Ridge Qb)

```python
def gaussians(x):
    scale = 0.1
    results = []
    for mean in np.linspace(0, 1, 20):
        results += [
            np.exp(-((x-mean)*(x-mean))/(2*scale*scale))
        ]
    return(results)
```

```python
Xs = np.linspace(-0.3, 1.3, 200)
gaussians_array =[]
for i in range(20):
    gaussians_array += [[]]
for x in Xs:
    for i in range(len(gaussians(x))):
        gaussians_array[i] += [gaussians(x)[i]]
for i in range(20):
    plt.plot(Xs,gaussians_array[i],label = 'Gaussian Plot')
```

Overfitting Study

```python
training_error_array = []
testing_error_array = []
lamdas = []
for param in tqdm.tqdm(np.logspace(-5, 0,20)):
    lamdas+=[param]

    training_error_order = 0
    testing_error_order = 0
    for one_out in range(25):
        X_train = np.reshape(list(np.linspace(0, 0.9, N)[0:one_out])+list(np.linspace(0, 0.9, N)[one_out+1:N]), (N-1, 1))
        X_test = np.reshape(X[one_out], 1)
        y_train = np.cos(10*X_train**2) + 0.1 * np.sin(100*X_train)
        y_test = np.cos(10*X_test**2) + 0.1 * np.sin(100*X_test)


        trigo1 = []
        for i in range(len(X_train)):
            line = [1.0]
            for m in range(20):
                line += [gaussians(X_train[i][0])[m]]
            trigo1 += [line]
        
        trigo1 = np.matrix(trigo1)

        trigo1_pred = []
        for i in range(len(X)):
            line = [1.0]
            for m in range(20):
                line += [gaussians(X[i][0])[m]]
            trigo1_pred += [line]
        
        trigo1_pred = np.matrix(trigo1_pred)

        inv1 = np.linalg.inv(np.dot(trigo1.T,trigo1) + (param * np.identity(21)) )
        weight1 = np.dot(np.dot(inv1,trigo1.T),y_train)
        mean1 = np.array(np.dot(weight1.T,trigo1_pred.T).T)
        training_error = 0
        testing_error = 0
        for i in range(len(mean1)):
            if i == one_out:
                testing_error += (mean1[i][0]-np.array(y)[i][0])**2
            else:
                training_error += (mean1[i][0]-np.array(y)[i][0])**2
        training_error = training_error/24.0
        training_error_order += training_error/25.0
        testing_error_order += testing_error/25.0
    
    training_error_array += [training_error_order]
    testing_error_array += [testing_error_order]
```

```python
plt.figure(4, figsize=(16, 10))
plt.title('Influence of Lambda over Training and Testing Errors')
plt.plot(lamdas,training_error_array,label = 'Training Error')
plt.plot(lamdas,testing_error_array,label = 'Testing Error')
plt.legend()
plt.xlabel('Lambda')
plt.xscale('log')
#plt.axvline(x=0.02)
```

It looks like the optimal value for lambda is 0.02. <br>
If lambda >> 0.02 => Underfitting (ex:10) <br>
If lambda << 0.02 => Overfitting (ex:0.000001) <br>

```python
trigo1 = []
for i in range(len(X)):
    line = [1.0]
    for m in range(20):
        line += [gaussians(X[i][0])[m]]
    trigo1 += [line]
        
trigo1 = np.matrix(trigo1)

trigo1_pred = []
for i in range(len(x_pred)):
    line = [1.0]
    for m in range(20):
        line += [gaussians(x_pred[i])[m]]
    trigo1_pred += [line]
        
trigo1_pred = np.matrix(trigo1_pred)

inv_under = np.linalg.inv(np.dot(trigo1.T,trigo1) + (10.0 * np.identity(21)) )
weight_under = np.dot(np.dot(inv_under,trigo1.T),y)
mean_under = np.array(np.dot(weight_under.T,trigo1_pred.T).T)

inv_over = np.linalg.inv(np.dot(trigo1.T,trigo1) + (0.000000000000001 * np.identity(21)) )
weight_over = np.dot(np.dot(inv_over,trigo1.T),y)
mean_over = np.array(np.dot(weight_over.T,trigo1_pred.T).T)

inv_fit = np.linalg.inv(np.dot(trigo1.T,trigo1) + (0.02 * np.identity(21)) )
weight_fit = np.dot(np.dot(inv_fit,trigo1.T),y)
mean_fit = np.array(np.dot(weight_fit.T,trigo1_pred.T).T)
```

```python
plt.figure(5, figsize=(16, 10))
plt.scatter(X,Y,color='darkred',label = 'data')
plt.plot(x_pred,mean_under,label = 'Underfitting (Lambda:10)')
plt.plot(x_pred,mean_over,label = 'Overfitting (Lambda:10e-15)')
plt.plot(x_pred,mean_fit,label = 'Reasonable Fit (Lambda:0.02)')
plt.title('Effect of Lambda over Regression with Gaussian basis functions')
plt.ylim(-2.,5.)
plt.legend()
```
